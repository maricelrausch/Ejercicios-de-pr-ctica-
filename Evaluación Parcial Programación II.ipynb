# Codigo

````python
#definir clase biblioteca con método constructor
class Biblioteca:
    def __init__(self, inventario):
        self.inventario = inventario 
        
    #definir los diferentes métodos (prestamos, agregar libros, disponibles)
    def agregar_libro(self, titulo, autor, año, estado):
        libro = {"titulo": titulo, "autor": autor, "año": año, "estado": "Disponible"}
        self.inventario.append(libro)
    def prestar_libro(self, titulo, usuario):
        for libro in self.inventario:
            #utilizo metodo if para que si un libro esta disponible y es prestado, cambie su estado
            if libro["titulo"] == titulo:
                if libro["estado"] == "Disponible":
                    libro["estado"] = "Prestado"
                    #utilizo un print para mostrar una alerta de prestamo 
                    print (f"Se prestó el libro {titulo} a {usuario}")
                    return
                else:
                    #print que avise que determinado libro no esta disponible
                    print(f'El libro no está disponible')
                    return
    def libros_disponibles(self):
        #utilizo un for para recorrer la lista de libros y mostrar los libros disponibles
        disponibles = [libro for libro in self.inventario if libro["estado"] == "Disponible"]
        print("estan disponibles estos libros: ")
        for libro in disponibles:
            #utilizo un print para mostrar los libros disponibles, iterandolos con un for nuevamente. 
            print(f'- {libro["titulo"]} de {libro["autor"]} ({libro["año"]})')


#determino una variable que contenga los diccionarios proporcionados por el docente
libros_iniciales = [
    {"titulo": "Cien Años de Soledad", "autor": "Gabriel García Márquez", "año": 1967, "estado": "Disponible"},
    {"titulo": "1984", "autor": "George Orwell", "año": 1949, "estado": "Disponible"},
    {"titulo": "El Principito", "autor": "Antoine de Saint-Exupéry", "año": 1943, "estado": "Disponible"},
]

#crear instancia para agregar el diccionario anterior al inventario
biblioteca = Biblioteca(libros_iniciales)

#comprobar si funciona el codigo llamando al método de prestamo. 
biblioteca.prestar_libro("1984", "Juan Kunzebich")
#comprobar si funciona el codigo llamando al método de agregar libros
biblioteca.agregar_libro("Métodos","Maria Rodriguez",1918,"Disponible")
#comprobar como queda el listado de libros después de los cambios realizados
biblioteca.libros_disponibles()

````
